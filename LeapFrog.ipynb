```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
import time
```

```python
# Define the LinkedInBot class
class LinkedInBot():
    # Define the init method
    def __init__(self, email, password, company, position):
        # Define the path to the Chrome driver - Enter the location of your chromedriver.exe file
        chromedriver_loc = '/Users/username/Downloads/chromedriver.exe'
        # Set the service and options for the Chrome driver
        service = Service(executable_path=chromedriver_loc)
        # Options for Chrome driver
        options = webdriver.ChromeOptions()
        # Create a new instance of the Chrome driver
        self.driver = webdriver.Chrome(service=service, options=options)

        # Navigate to the LinkedIn login page
        # 1 keyword search - Examples: Analyst, Recruiter, Manager, etc.
        # self.driver.get('https://www.linkedin.com/company/{your company - eg. Apple}/people/?keywords={your position - eg. Recruiter}')
        # 2 keyword locate - Examples: Data Analyst, Data Engineer, Senior Recruiter, etc.
        # self.driver.get('https://www.linkedin.com/company/{your company - eg. Apple}/people/?keywords={keyword 1 - eg. data}%20{keyword 2 - eg. analyst}')
        website = f"https://www.linkedin.com/company/{company}/people/?keywords={position}"
        self.driver.get(website)
        time.sleep(5)

        # Enter the email and password and click the "Sign in" button
        email_field = self.driver.find_element(By.ID, 'username')
        email_field.send_keys(email)
        time.sleep(3)
        password_field = self.driver.find_element(By.ID, 'password')
        password_field.send_keys(password)
        time.sleep(2)
        signin_button = self.driver.find_element(By.XPATH, "//button[@type='submit']")
        signin_button.click()
        time.sleep(4)
    
    # Define method to locate links
    def locate_links(self):
        # Find the unordered list element
        ul_element = self.driver.find_element(By.XPATH, "//ul[@class='display-flex list-style-none flex-wrap']")

        # Find all the a elements within the li elements of the unordered list and extract their href attributes
        hrefs = [a.get_attribute("href") for a in ul_element.find_elements(By.XPATH, ".//li//div[2]//a")]

        # Remove duplicates from the hrefs list
        link_list = list(set(hrefs))

        # Close the browser window
        self.driver.quit()

        return link_list
    
```

```python
# Enter your email and password here
email, password = "your_email@gmail.com", "your_password"
# Create a new instance of the bot - Enter your desired company name and position
bot = LinkedInBot(email, password, "Enter Company Name", "Enter Position")
# Locate all the profile links
links_located = bot.locate_links()
# Print the profile links
print(links_located)
```

```python
class getlinks():
    def __init__(self, email, password):
            # Define the path to the Chrome driver - Enter the location of your chromedriver.exe file
            chromedriver_loc = '/Users/username/Downloads/chromedriver.exe'
            # Define the path to the Chrome driver
            service = Service(executable_path=chromedriver_loc)
            # Options for Chrome driver
            options = webdriver.ChromeOptions()
            # Create a new instance of the Chrome driver
            self.driver = webdriver.Chrome(service=service, options=options)

            # Navigate to the LinkedIn login page
            self.driver.get('https://www.linkedin.com/login')

            # Enter the email and password and click the "Sign in" button
            email_field = self.driver.find_element(By.ID, 'username')
            email_field.send_keys(email)
            time.sleep(3)
            password_field = self.driver.find_element(By.ID, 'password')
            password_field.send_keys(password)
            time.sleep(4)
            signin_button = self.driver.find_element(By.XPATH, "//button[@type='submit']")
            signin_button.click()
            time.sleep(3)
    def linkgetter(self, links, premium=True, message= None):
        for link in links:
            # Navigate to the profile page
            try:
                # Navigate to the profile page
                self.driver.get(link)
                time.sleep(3)
                # Check if
                # Connect button exists on profile homepage
                try:
                    # Find connect button
                    connect_button = self.driver.find_element(By.XPATH, "(//li-icon[@type='connect'])[2]")
                    # Click connect button
                    connect_button.click()
                    time.sleep(3)
                    # Find send button without note
                    send_button = self.driver.find_element(By.XPATH, "//button[@aria-label='Send now']")
                    # Click send button without note
                    send_button.click()
                    time.sleep(3)
                # Else 
                # find connect button in the more dropdown
                except ElementClickInterceptedException:
                    # Find more button to view dropdown
                    more_button = self.driver.find_element(By.XPATH, "(//button[@aria-label='More actions'])[2]")
                    # Click more button
                    more_button.click() 
                    time.sleep(3)
                    # Check if already connected or not
                    try: # If not connected 
                        # Find connect button
                        more_connect_button = self.driver.find_element(By.XPATH, "(//div[@class='artdeco-dropdown__content-inner'])[2]//ul//li[3]//span") 
                        # locate more button and click
                        more_connect_button.click()
                        time.sleep(3)
                        # If premium send with note
                        if premium:
                            add_note = self.driver.find_element(By.XPATH, "(//button[@aria-label='Add a note'])[1]") 
                            add_note.click()
                            time.sleep(3)
                            # add message
                            message_field = self.driver.find_element(By.XPATH, "(//textarea[@id='custom-message'])[1]")
                            message_field.send_keys(message)
                            time.sleep(3)
                            # send connection request
                            send_connect = self.driver.find_element(By.XPATH, "(//button[@aria-label='Send now'])[1]") 
                            send_connect.click()
                            time.sleep(3)
                        # Else send without note
                        else:
                            send_connect = self.driver.find_element(By.XPATH, "(//button[@aria-label='Send now'])[1]") 
                            send_connect.click()
                            time.sleep(3)
                    # Else if already connected 
                    except NoSuchElementException:
                        # Already connected
                        print("Already connected with this person. Skipping...")
                        continue
            # Else if profile is not available
            except ElementClickInterceptedException:
                # Profile not available
                print("Profile not available. Skipping...")
                continue
        # Close the browser window
        self.driver.quit()
```
